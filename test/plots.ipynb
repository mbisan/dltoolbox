```python
import dltoolbox.activationFunctions as af

import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{amsmath}\usepackage{amsfonts}')

def plot_soft_relu():
    fig, axs = plt.subplots(1)

    linspace = np.arange(-5,5,0.01)

    soft_relu = np.log(1+np.exp(linspace))
    sigmoid = af.sigmoid.f(linspace)
    step = 1* (linspace>0)



    axs.set_title(r"Softplus")
    axs.set(ylabel = r"$f(x) = \log(1+e^x)$, $f'(x)=\frac{1}{1+e^{-x}}$")
    axs.set(xlabel = r"$x$")
    axs.plot(linspace, soft_relu, linspace, sigmoid)

def plot_esponentzialki_haztatutako_bb():
    linspace = np.arange(1,50,1)

    beta_09 = (0.1)* (0.9**linspace)
    beta_08 = (0.2)* (0.8**linspace)
    beta_03 = (0.7)* (0.3**linspace)

    fig, axs = plt.subplots(1)
    axs.set(ylabel = r"$(1-\beta)\beta^n$")
    axs.set(xlabel = r"$n$")
    axs.plot(linspace, beta_09, "*", linspace, beta_08, "bs", linspace, beta_03, "g^")
    axs.legend([r'$\beta = 0.9$', r'$\beta = 0.8$', r'$\beta = 0.3$'])
    plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['figure.dpi'] = 150
```

```python
plot_soft_relu()
```

```python
plot_esponentzialki_haztatutako_bb()
```
